```python
matrix = [
    [2,3,5,4],
    [1,4,7,3],
    [2,5,6,5],
    [2,3,4,1]
]

# matrix = [
#     [2,5],
#     [3,4]
# ]

# matrix = [
#     [4, 10, 8],
#     [5, 10, 7],
#     [6, 7, 8]
# ]

print(matrix)
from pprint import pprint
from queue import PriorityQueue

def getNeighbors(matrix, x,y):
    nb = []
    if x + 1 < len(matrix):
        nb.append([x+1,y])
    if x - 1 >= 0:
        nb.append([x-1,y])  
    if y + 1 < len(matrix[0]):
        nb.append([x,y+1])
    if y - 1 >= 0:
        nb.append([x,y-1])
    return nb

def getKey(x,y):
    return str(x) + '-' + str(y)
def solve(matrix, point):
    x = point[0]
    y = point[1]
    
    hashMap = dict()
    queue = PriorityQueue()
    queue.put([matrix[x][y], [x,y]])
    hashMap[getKey(x,y)] = [0,None]
    
    
    currentMaxNode = [0,0]
    maxPath = 0
    
    while not queue.empty():
        val, node = queue.get()
        neighbors = getNeighbors(matrix, node[0], node[1])
        for n in neighbors:
            x_n = n[0]
            y_n = n[1]
            if matrix[x_n][y_n] < matrix[node[0]][node[1]]:
                newVal = hashMap[getKey(node[0],node[1])][0] + 1
                if getKey(x_n, y_n) in hashMap:
                    value = hashMap[getKey(x_n, y_n)] # current val
                    if value[0] < newVal: # update new val and parent
                        hashMap[getKey(x_n, y_n)] = [newVal, [node[0], node[1]]]
                else:
                    hashMap[getKey(x_n, y_n)] = [newVal, [node[0], node[1]]]
                if hashMap[getKey(x_n, y_n)][0] > maxPath:
                    maxPath = hashMap[getKey(x_n, y_n)][0]
                    currentMaxNode = [x_n, y_n]
                queue.put([matrix[x_n][y_n], [x_n,y_n]])
    pprint(hashMap)
    newMatrix = [[[0] *len(matrix[0])]  * len(matrix)]
    print(newMatrix)
    print("curr",currentMaxNode)
    parent = currentMaxNode
    path = []
    while parent != None:
        path.append(hashMap[getKey(parent[0], parent[1])][1])
        parent = hashMap[getKey(parent[0], parent[1])][1]
    print(path)
        
    
    
solve(matrix, [2,2])    
    
    
```
